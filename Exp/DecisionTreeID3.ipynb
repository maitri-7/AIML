```python
from sklearn.tree import plot_tree, DecisionTreeClassifier,export_graphviz, export_text
from sklearn.datasets import load_wine
import matplotlib.pyplot as plt
import graphviz
plt.figure(figsize=(60, 20))
wine = load_wine()
clf = DecisionTreeClassifier(criterion='entropy')
clf.fit(wine.data, wine.target)
plot_tree(clf,feature_names=wine.feature_names,
filled=True,class_names=wine.target_names,rounded=True)
plt.title("Decision tree trained on all the iris features")

plt.show()
r = export_text(clf, feature_names=wine.feature_names, )
print(r)
```
