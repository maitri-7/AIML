```python
import pandas as pd
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sns
input_file = 'https://raw.githubusercontent.com/51stDimension/AIML/main/Data/EXP1.csv'
data = pd.read_csv(input_file)
data = data.drop('Unnamed: 0',axis=1)
data.head()
print("Attributes are:")
data.columns
print("Target Hypothesis")
target = ['Sunny', '?', 'High', 'Strong', '?']
print(target)
num = 1
col = data.shape[1]
hypo = []
hypo = ['%' for i in range(col-2)]
print('Initial hypothesis: ')
print(hypo)
print('\n')
for row in data.iterrows():
  print('Reading row number: ' + str(num))
  outcome = row[1]['Yes']
  if outcome == 'Yes':
    print('Positive outcome, updating hypothesis')
    for x in range(0,col-2):
      if hypo[x]=='%':
        hypo[x] = row[1][x]
      elif hypo[x]=='?':
        continue
      elif hypo[x]!=row[1][x]:
        hypo[x] = '?'
    print('Updated Hypothesis')
    print(hypo)
    print('\n')

  else:
    print('Negative outcome, skipping data')
    print('\n')
  num += 1

print('Final Hypothesis')
print(hypo)
```
